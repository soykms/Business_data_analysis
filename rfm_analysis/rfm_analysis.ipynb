```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# data load
df=pd.read_csv('rfm_mart.csv')
df
```

```python
# data EDA
df.info()
df.isnull().sum()
```

```python
# Logistic regression
x_columns=['recency','frequency','monetary']
df[x_columns]

y_target=['is_back']
df[y_target]
```

```python
# Test Train split 
# help(train_test_split)
X_train, X_test, y_train, y_test = train_test_split(df[x_columns], df[y_target], test_size=0.2, random_state=42)
```

```python
# Model train
model=LogisticRegression()
model.fit(X_train,y_train)

help(LogisticRegression)
```

```python
# Model validation
model.score(X_test,y_test)
```

```python
# coefficient
coef = pd.DataFrame({'features':df[x_columns].columns, 'coefficient':model.coef_[0]})
coef
```

```python
# Percentitle score - recency

np.percentile(df['recency'],[20,40,60])
a1,a2,a3= np.percentile(df['recency'],[20,40,60])
```

```python
def percent_recency(x) : 
    if x <= a1:
        return 4
    elif x > a1 and x <=a2:
        return 3
    elif x > a2 and x <a3:
        return 2
    elif x >= a3:
        return 1
```

```python
df['recency_scroe']=df['recency'].apply(percent_recency)*coef.iloc[0,-1]*(-1)
df
```

```python
# Percentitle score - frequency

np.percentile(df['frequency'],[10,50,80])
b1,b2,b3= np.percentile(df['frequency'],[10,50,80])
```

```python
def percent_frequency(x) : 
    if x <= b1:
        return 1
    elif x > b1 and x <=b2:
        return 3
    elif x > b2 and x <b3:
        return 2
    elif x >= b3:
        return 1
```

```python
df['frequency_scroe']=df['frequency'].apply(percent_frequency)*coef.iloc[1,-1]
df
```

```python
# Percentitle score - monetary

np.percentile(df['monetary'],[20,40,80])
c1,c2,c3= np.percentile(df['monetary'],[20,40,80])
```

```python
def percent_frequency(x) : 
    if x <= c1:
        return 1
    elif x > c1 and x <=c2:
        return 3
    elif x > c2 and x <c3:
        return 2
    elif x >= c3:
        return 1
```

```python
df['monetary_scroe']=df['monetary'].apply(percent_frequency)*coef.iloc[2,-1]
df

# 회귀계수랑 임의로 나눈 그룹 스코어랑 값하는게 논리적으로 맞나?
```

```python
# total score
df['total_score']=df['frequency_scroe']+df['monetary_scroe']+df['recency_scroe']
df
```

```python
# grouping
# 의사결정을 해가면서 기준을 정해야함
# 분포를 먼저 한번 보고 하면 좋을듯
t1, t2, t3, t4=np.percentile(df['total_score'],[20,50,70,80])
t1, t2, t3, t4
```

```python
def level(x):
    if x<= t1 :
        return 5
    elif x> t1 and x <=t2:
        return 4
    elif x> t2 and x <=t3:
        return 3
    elif x> t3 and x <=t4:
        return 2
    elif x > t4:
        return 1
```

```python
df['level']=df['total_score'].apply(level)
```

```python
df
```

```python
# retention

pivot = df.groupby('level').agg({'is_back':'sum', 'mem_no':'count'})
pivot

pivot['retention']=pivot.is_back/pivot.mem_no
pivot
```
